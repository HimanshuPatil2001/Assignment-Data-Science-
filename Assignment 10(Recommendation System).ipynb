```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
book = pd.read_csv('book.csv',encoding = "latin1")
book
```

```python
book.rename(columns={'User.ID': 'ID',
                   'Book.Title': 'Title',
                    'Book.Rating' : 'Rating'},
          inplace=True,)
book
```

```python
book.drop('Unnamed: 0',axis=1,inplace=True)
book
```

```python
book_df = book.pivot_table(index = 'ID',
                    columns = 'Title',
                    values = 'Rating').reset_index(drop=True)
book_df
```

```python
book_df.fillna(0,inplace = True)
book_df
```

```python
sim = 1- pairwise_distances(book_df.values,metric = 'cosine')
sim
```

```python
np.fill_diagonal(sim,0)
sim
```

```python
Similarity = pd.DataFrame(sim)
Similarity
```

```python
Similarity.index = book['ID'].unique()
Similarity.columns = book['ID'].unique()
Similarity
```

```python
Similarity.idxmax(axis=1)
```

```python
Similarity.loc[162121].idxmax()
```

```python
def recommend(custID):
    simID = Similarity.loc[custID].idxmax()
    simID_books = book[book['ID'] == simID].Title
    custID_books = book[book['ID'] == custID].Title
    return set(simID_books) - set(custID_books)
```

```python
recommend(162121)
```

```python
recommend(3546)
```
