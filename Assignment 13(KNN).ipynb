```python
import pandas as pd
from sklearn.model_selection import KFold, cross_val_score, GridSearchCV
from sklearn.neighbors import KNeighborsClassifier
import matplotlib.pyplot as plt
import numpy as np
```

### ZOO

```python
data = pd.read_csv('Zoo.csv')
data
```

```python
data.isnull().sum()
```

##### This data frame has no null elements

```python
X = data.iloc[:,1:-1]
type = data.iloc[:,-1]
```

```python
X
```

```python
type
```

```python
n_neighbour = np.array(range(1,41))
params = {'n_neighbors' : n_neighbour}
model = KNeighborsClassifier()
```

```python
grid = GridSearchCV(estimator = model,param_grid = params)
grid.fit(X,type)
```

```python
print(grid.best_params_)
```

```python
print(grid.best_score_)
```

```python
final_model = KNeighborsClassifier(n_neighbors = 1)
final_model.fit(X,type)
```

```python
final_model.predict(X)
```

```python
Precision = pd.DataFrame({'Predict':final_model.predict(X),
                        'Actual':type})
Precision
```

```python
kfold = KFold(n_splits = 10)
result =cross_val_score(final_model,X,type,cv = kfold)
result
```

```python
result.mean()
```

### GLASS

```python
data = pd.read_csv('glass.csv')
data
```

```python
data.isnull().sum()
```

```python
X = data.iloc[:,:-1]
Y = data.iloc[:,-1]
X
```

```python
Y
```

```python
n_neighbor = np.array(range(1,41))
params = {'n_neighbors':n_neighbor}
model = KNeighborsClassifier()
grid = GridSearchCV(estimator = model,param_grid= params)
```

```python
grid.fit(X,Y)
```

```python
print(grid.best_params_)
print(grid.best_score_)
```

```python
final_model = KNeighborsClassifier(n_neighbors = 1)
final_model.fit(X,Y)
```

```python
final_model.predict(X)
```

```python
Precision = pd.DataFrame({'Prediction':final_model.predict(X),
                         'Actual':Y})
Precision
```

```python
kfold = KFold(n_splits = 10)
result = cross_val_score(final_model,X,Y,cv = kfold)
result
```

```python
result.mean()
```
