```python
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules
from mlxtend.preprocessing import TransactionEncoder
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('book.csv')
data
```

```python
data.shape
```

##### We will see the difference for min_support of 0.1 ,0.2 ,0.3 .

```python
frequent_item = apriori(data,min_support = 0.1, use_colnames = True)
frequent_item
```

```python
frequents_items = apriori(data,min_support = 0.2,use_colnames = True)
frequents_items
```

```python
frequent_items = apriori(data,min_support = 0.3, use_colnames = True)
frequent_items
```

#### We will use min_support value of 0.1

```python
frequent_item
```

##### We will try for 3 different values of 0.1 ,0.5 ,0.7

```python
rules = association_rules(frequent_item,metric = 'lift',min_threshold = 0.1)
rules
```

```python
rules.sort_values('lift',ascending = False)
```

```python
rules = association_rules(frequent_item,metric = 'lift',min_threshold = 0.5)
rules
```

```python
rules.sort_values('lift',ascending = False)
```

```python
rules = association_rules(frequent_item,metric = 'lift',min_threshold = 0.7)
rules
```

```python
rules.sort_values('lift',ascending = False)
```

##### Here you can see the difference for lift as a feature

#### We will use the rules with threshold values of 0.7

```python
rules
```

```python
rules[['support','confidence','lift']].hist()
```

```python
plt.scatter(rules['support'],rules['confidence'])
plt.show()
```
